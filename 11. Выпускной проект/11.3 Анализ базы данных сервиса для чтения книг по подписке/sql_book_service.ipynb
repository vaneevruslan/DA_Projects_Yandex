# Исследовательский анализ данных

## Считаем кол-во выпущенных книг после 1 января 2000 года

```python
sql_query_1 = """
                  SELECT COUNT(*) AS books_number
                  FROM books
                  WHERE publication_date::date > '2000-01-01'
                  
              """

query(sql_query_1)
```

**Вывод:** 

После 1 января 2000 года вышло `819` книг.

## Считаем кол-во обзоров и среднюю оценку для каждой книги 

```python
sql_query_2 = """
                  SELECT b.title,
                         COUNT(DISTINCT rv.review_id) AS reviews_number,
                         ROUND(AVG(rt.rating),2) AS avg_rating
                  FROM books AS b
                  LEFT JOIN reviews AS rv ON b.book_id = rv.book_id
                  LEFT JOIN ratings AS rt ON b.book_id = rt.book_id
                  GROUP BY 1
                  ORDER BY 2 DESC
                  
              """

query(sql_query_2)
```

**Вывод:**

Больше всего обзоров у книги `Memoirs of a Geisha` - `8`, далее идет `Twilight` - `7` и замыкает первую тройку `Harry Potter and the Chamber of Secrets` с `6` обзорами.

## Находим издательство, которое выпустило наибольшее число книг, толще 50 страниц

```python
sql_query_3 = """
                  SELECT p.publisher,
                         COUNT(b.book_id) AS books_number
                  FROM books AS b
                  LEFT JOIN publishers AS p ON b.publisher_id = p.publisher_id
                  WHERE b.num_pages > 50
                  GROUP BY 1
                  ORDER BY 2 DESC
                  LIMIT 1
                  
              """

query(sql_query_3)
```

**Вывод:**

Издательство, которое выпустило наибольшее число книг толще 50 страниц - `Penguin Books`(42 книги).

## Находим автора с самой высокой средней оценкой книг

```python
sql_query_4 = """ 
                    SELECT author,
                           ROUND(AVG(avg_rating), 2) AS avg_rating
                    FROM
                      (SELECT a.author,
                              b.book_id,
                              ROUND(AVG(rt.rating), 2) AS avg_rating
                       FROM books AS b
                       LEFT JOIN ratings AS rt ON b.book_id = rt.book_id
                       LEFT JOIN authors AS a ON b.author_id = a.author_id
                       GROUP BY 1,
                                2
                       HAVING COUNT(rt.rating) >= 50
                       ) AS books_with_50_ratings
                    GROUP BY 1
                    ORDER BY 2 DESC
                    LIMIT 1    
             """

query(sql_query_4)
```

**Вывод:**

`J.K. Rowling/Mary GrandPré` - автор с самой высокой средней оценкой книг (с 50 и более оценками) и в среднем составляет - `4.29`.

## Находим среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
sql_query_5 = """
                 WITH num_of_reviews_per_user AS
                     (SELECT username,
                            COUNT(review_id) AS reviews_number
                     FROM reviews
                     WHERE username IN (SELECT username
                                        FROM ratings
                                        GROUP BY 1
                                        HAVING COUNT(rating) > 50)
                     GROUP BY 1)
                  
                 SELECT ROUND(AVG(reviews_number))::integer AS avg_reviews
                 FROM num_of_reviews_per_user
              """

query(sql_query_5)
```

**Вывод:**

Cреднее количество обзоров от пользователей, которые поставили больше 50 оценок составляет - `24`
